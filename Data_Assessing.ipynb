```python
# What is Data assessing?

# Data assessing means checking data quality before analysis.
# Types of unclean Data
# There are two types  of unclean data
    # 1. Dirty Data (with quality issues): Also known low quality data. Has content issues
    # 2. Messy Data (with tidiness issues): Also known as untidy data. Has structural issues

# Dirty Data: When the data has following kind of issues called dirty data
    # 1. Duplicated Data
    # 2. Missing Data
    # 3. Corrupt Data
    # 4. Inaccurate Data
# Messy Data: When the data has structural issue is it known as messy or untidy data
    # Tidy data has the following properties:
        # 1. Each variables form a column
        # 2. Each observation forms a row
        # 3. Each observational unit forms a table
    # If a data violates any one of the above it is know as untidy data

# Type of Assessment: There 2 types of assessment
    # 1. Manual-Looking through the data manually in googles sheets
    # 2. Programmatic- by using pandas function susc as info(), describe(), sample()

# Steps in Assesment: 
  # There are two step involved in data assessment 
    # Discover
    # Document
```
