```python
# Import required libraries
import pandas as pd
```

```python
# Import dataset
patients=pd.read_csv('patients.csv')
treatment=pd.read_csv('treatments.csv')
adverse=pd.read_csv('adverse_reactions.csv')
```

```python
# Copy dataset
patients_copy=patients.copy()
treatment_copy=treatment.copy()
adverse_copy=adverse.copy
```

```python
# Completeness issues
    # Missing values for contact details of 12 patients
    # Only 280 rows are there instead of 350
    # hba1_c change col has missing values
```

```python
# Define
    # Substract hba1c_Start from hba1c_end to get hba1c_change
```

```python
# Code
treatment_copy['hba1c_change']=treatment_copy['hba1c_start']-treatment_copy['hba1c_end']
```

```python
# Test
treatment_copy['hba1c_change'].isnull().sum()
```

```python
# Check basic information
treatment_copy
```

```python
# Tidiness Issues
    # Phone +email in contact col patients
    # Merge the adverse table to treatment table
    # Novodra and aurolin col treatment
```

```python
# Define
# create different column from aurolin and novodra
```

```python
# Code
treatment_copy=pd.melt(treatment_copy, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end'],
                       var_name='treatment', value_name='dose') 
```

```python
treatment_copy
```

```python
# Remove [-] from dose column
treatment_copy=treatment_copy[treatment_copy['dose']!='-']
treatment_copy
```

```python
# Convert dose column into two column
treatment_copy[['start', 'end']] = (
    treatment_copy['dose']
    .astype(str)
    .str.split('-', n=1, expand=True)
)
```

```python
treatment_copy
```

```python
# Drop dose column
treatment_copy.drop(columns='dose', inplace=True)
```
