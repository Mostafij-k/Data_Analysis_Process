```python
            #------ API------#
# API means Application Programming Interface
# An API is a way for one software to ask another software for data or services.
# Most API return data in JSON format
# Client>API Request>Server
# Server>API Response>Client

# Gathering Data using API
import requests
import pandas as pd
response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=aa870f02d8bc0d3ec4cef56143ef0705&language=en-US&page=1').json()
response
for i in response['results']:
    print(i['title'], i['vote_average'])
```

```python
popularity=[]
vote_count=[]
original_language=[]
title=[]
vote_average=[]
overview=[]
release_date=[]
for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
d={'title':title, 'overview':overview, 'original_language': original_language,
    'release_date':release_date,'popularity':popularity, 'vote_count':vote_count, 'vote_average':vote_average}
df=pd.DataFrame(d)
df
```
