```python
# Fetch the pupular movie information and prepared dataframe

# Create empty dataframe
final_Data=pd.DataFrame()
# Get access the first 30 pages
for j in range (1, 31):
    response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=aa870f02d8bc0d3ec4cef56143ef0705&language=en-US&page={}'.format(j)).json()
# Create empty list and each list store one column of the dataframe
    popularity=[]
    vote_count=[]
    original_language=[]
    title=[]
    vote_average=[]
    overview=[]
    release_date=[]

    # Extract movie information which act as a feature
    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])
    # Create dictionary     
    d={'title':title, 'overview':overview, 'original_language': original_language,
        'release_date':release_date,'popularity':popularity, 'vote_count':vote_count, 'vote_average':vote_average}
    # Convert dictionary to Dataframe
    df=pd.DataFrame(d)
    # Join old dataframe with new dataframe
    final_Data = pd.concat([final_Data, df], ignore_index=True)
    
# Check Dataframe shape
final_Data.shape
```

```python
# Create CSV file and save to present directory
final_Data.to_csv("Popular_movies.csv")
```

```python
# Another way

# Import to required packages
import requests  # Fetch data from TMDB API
import pandas as pd    # write data into CSV

# Create Dataset
movies=[]
for page in range(1, 31):
    url = f"https://api.themoviedb.org/3/movie/popular?api_key=aa870f02d8bc0d3ec4cef56143ef0705&language=en-US&page={page}"
    response=requests.get(url).json()
    for m in response['results']:
        movies.append({
            'title': m['title'],
            'overview': m['overview'],
            'original_language': m['original_language'],
            'release_date': m['release_date'],
            'popularity': m['popularity'],
            'vote_count': m['vote_count'],
            'vote_average': m['vote_average']
        })

final_Data = pd.DataFrame(movies)
final_Data.shape
```
